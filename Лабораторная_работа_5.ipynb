```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/preprocessed_data_soobshchestva_itmo.csv")
df.head()
```

```python
#—Å–æ—Ö—Ä–∞–Ω—è–µ–º —Å—Ç–æ–ª–±–µ—Ü processed_text –≤ –º–∞—Å—Å–∏–≤ —Å—Ç—Ä–æ–∫
db = []
for index, row in df.iterrows():
    db.append(row['processed_text'])
#—É–¥–∞–ª—è–µ–º –≤—Å–µ —Ç–∏–ø—ã –∫—Ä–æ–º–µ str
corpus = [text for text in db if isinstance(text, str) and text != 'nan']
corpus
```

```python
#BoW
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#–≤–µ–∫—Ç–æ—Ä–∏–∑—É–µ–º —Å—Ä–∞–∑—É –≤–µ—Å—å –∫–æ—Ä–ø—É—Å
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus).toarray()
X
```

```python
#–≤ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ç–æ—Ä –ø–æ–¥–∞–µ–º –Ω–æ–≤—ã–π —Ç–µ–∫—Å—Ç
vectorizer.transform(['–∫–∞–∑–∞—Ç—å—Å—è','–≤—ã—Ö–æ–¥–Ω–æ–π','—Ç—Ä—É–¥–æ–≤–æ–π','–±—É–¥–Ω–∏','–≤—Ä—ã–≤–∞—Ç—å—Å—è','–ø–∞—Ä–∞','–¥–µ–Ω—å','—Å–≤—è–∑—å','–≤–æ–ø—Ä–æ—Å','–º–µ–º']).toarray()
```

```python
#—Å–æ—Ö—Ä–∞–Ω—è–µ–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä –ø—É—Ç–µ–º BoW
bow_df = pd.DataFrame(X)
bow_df.to_csv('/content/drive/MyDrive/bow_vectors.csv', index=False)
```

```python
#TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
X_2 = vectorizer.fit_transform(corpus).toarray()
X_2
```

```python
#—Å–æ—Ö—Ä–∞–Ω—è–µ–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä –ø—É—Ç–µ–º TF-IDF
TFIDF_df = pd.DataFrame(X_2)
TFIDF_df.to_csv('/content/drive/MyDrive/TF-IDF_vectors.csv', index=False)
```

```python
#Word2Vec
import gensim.downloader as api
```

```python
#–∑–∞–≥—Ä—É–∑–∫–∞ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω—É—é –º–æ–¥–µ–ª—å
wv = api.load('word2vec-ruscorpora-300')
```

```python
#—Å–æ–∑–¥–∞–Ω–∏–µ —Å–ø–∏—Å–∫–∞ —Å–ª–æ–≤ –±–µ–∑ —á–∞—Å—Ç–µ–π —Ä–µ—á–∏
vocab = []
for w in list(wv.index_to_key):
  vocab.append(w.split('_')[0])
```

```python
vectors_text = []
for text in corpus:
  vectors_word = []
  words = text.split()
  for i in words:
    if i in vocab:
      vectors_word.append(wv[list(wv.index_to_key)[vocab.index(i)]])
    else:
      vectors_word.append(np.zeros(300))
  vectors_text.append(np.mean(vectors_word, axis = 0))
```

```python
#–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–µ–∫—Ç–æ—Ä–æ–≤
len(vectors_text)
```

```python
#–ü—Ä–∏–º–µ—Ä –≤—ã–≤–æ–¥–∞ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ —Ä–∞–∑–º–µ—Ä–æ–º 300 —á–∏—Å–µ–ª
for i in range(2):
    print(f"–í–µ–∫—Ç–æ—Ä {i+1}:")
    print(vectors_text[i])
    print()
```

```python
# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤ –≤ —Ñ–∞–π–ª .npy
np.save('/content/drive/MyDrive/word2vec_vectors.npy', vectors_text)
```
